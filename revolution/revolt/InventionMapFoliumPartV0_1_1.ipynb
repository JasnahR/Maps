```python
years = {}
for i in json:
    if i["год"] not in years.keys():
        years[i["год"]] = []
    years[i["год"]].append({})
    years[i["год"]][-1]["name"] = i["название"]
    years[i["год"]][-1]["country"] = i["страна"]
    years[i["год"]][-1]["coordinates"] = i["coordinates"].replace("[", "").replace("]", "").replace(",", "").split()
    years[i["год"]][-1]["coordinates"][0] = float(years[i["год"]][-1]["coordinates"][0])
    years[i["год"]][-1]["coordinates"][1] = float(years[i["год"]][-1]["coordinates"][1])
    years[i["год"]][-1]["desc"] = i["описание"]
    years[i["год"]][-1]["icon"] = i["icon"]
    years[i["год"]][-1]["color"] = i["color"]
```
